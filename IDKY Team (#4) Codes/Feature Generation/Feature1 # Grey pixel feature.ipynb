###Identify the all degress of grey, count grey pixels and append to the previous DataFrame's column.

```
with open("/content/drive/Shared drives/MyDrive/Kaggle/clips(enhanced) test/_temp",'w') as f:
    f.write("25000")
```

```
with open("/content/drive/Shared drives/MyDrive/Kaggle/clips(enhanced) test/_temp",'r') as f:
    start=int(f.read())+1
print(f"Starting with {start} picture.")
data=pickle.load(open("/content/drive/Shared drives/MyDrive/Kaggle/Feature Generation (Enhanced)/Real size+grey feature.pkl",'rb'))
path="/content/drive/Shared drives/MyDrive/Kaggle/clips(enhanced) test/"
for i in tqdm(range(start,50001)):
    filepath=path+"clips-"+str(i)+".png"
    count=np.array(Image.open(filepath),dtype='bool')
    data.loc[i,"grey_pixel_number"]=int(count.sum()/3)
    if(i%50==0):
        with open("/content/drive/Shared drives/MyDrive/Kaggle/clips(enhanced) test/_temp",'w') as f:
            f.write(str(i))
    if(i%50==0):
        pickle.dump(data,open("/content/drive/Shared drives/MyDrive/Kaggle/Feature Generation (Enhanced)/Real size+grey feature.pkl",'wb'))
```
