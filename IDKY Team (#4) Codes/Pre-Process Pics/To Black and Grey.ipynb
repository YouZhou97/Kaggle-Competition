```
with open(path2+"_temp",'r') as f:
    start=int(f.read())+25000+1
print(f"\nStarting from Picture {start}")

def func(x):
    if(max(x)<210):
        if(max(x)-min(x)<10):
            return x
        elif(max(x)-min(x)<80):
            return np.array([x.mean()]*3,dtype='uint8')
    return np.array([255,255,255],dtype='uint8')

for m in tqdm(range(start,50001)):
    filepath=path+"clips-"+str(m)+".png"
    img=Image.open(filepath).convert('RGB')
    arr=np.apply_along_axis(func,2,np.array(img))
    #chop seperate dots.
    for i in range(256):
        j=24
        if(arr[i,j,0]!=255):
            if(arr[i-1:i+2,j-1:j+2].sum()>6400):
                arr[i,j]=np.array([255,255,255],dtype='uint8')
    arr=np.uint8(255)-arr
    img=Image.fromarray(arr,'RGB')
    img.save(open(path2+"clips-"+str(m)+".png",'wb'))

    if(m%100==0):
        with open(path2+"_temp",'w') as f:
            f.write(str(m))
    if(m%1000==0):
        print(f"\n{m} picture saved successfully!")
```
